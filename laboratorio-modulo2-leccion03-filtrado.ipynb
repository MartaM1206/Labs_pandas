```python
# Antes de empezar importamos la librería de NumPy para poder trabajar con todos sus métodos. 
import numpy as np
# ==============================================================================
# Tratamiento de datos
# ==============================================================================
import pandas as pd
# Ignorar warnings
# ==============================================================================
import warnings
warnings.filterwarnings("ignore")
# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
# Carga el dataset de trabajo
df_catalogo = pd.read_csv("datos/catalogo_completo.csv", index_col=0)
df_catalogo.head(1)
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
df_2021movies = df_catalogo.query('type == "Movie" and release_year == "2021"')
df_2021movies
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
df_90min = df_catalogo.query('duration == "90 min"').filter(items = ["title", "duration"])
df_90min
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
df_India_2018 = df_catalogo.query('country == "India" and release_year > 2018').filter(items = ["title", "release_year"])
df_India_2018
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
df_R_TV14 = df_catalogo.query('rating == "TV-14" or rating == "R"').filter(items = ["title", "rating"])
df_R_TV14
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
df_Nolan = df_catalogo.query('director == "Christopher Nolan"').filter(items = ["title"])
df_Scorsese = df_catalogo.query('director == "Martin Scorsese"').filter(items = ["title"])
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
lista_paises = ['United States', 'United Kingdom', 'Canada']
contiene = "/" ## Cuando hay más de un idioma se separan con '/'
filtro_idiomas = df_catalogo["Language"].str.contains(contiene, na = False)
```

```python
df_paises = df_catalogo[df_catalogo["country"].isin(lista_paises)]
df_varios = df_paises[filtro_idiomas]
df_varios
```

```python
print(f"{df_varios.shape[0]} títulos tienen más de un idioma")
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
filtro_lanzamiento = df_catalogo[df_catalogo["release_year"].between(2015, 2020, inclusive = "both" )]
filtro_lanzamiento.filter(["title", "release_year"])
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
filtro_puntos = df_catalogo[df_catalogo["IMDB Score"].between(6.5, 8.0, inclusive = "both" )]
df_ej8 = filtro_puntos[filtro_puntos["duration"].isnull()]
df_ej8
```

No se pueden aplicar a la vez un filtro de tipo float y otro de tipo bool, por lo que hay que aplicar uno de ellos y luego, sobre el df resultante de aplicar el primer filtro, aplicar el segundo. Aquí hemos aplicado between() para filtrar por las puntuaciones IMDB y, sobre el df resultante (filtro_puntos), hemos aplicado .isnull() sobre la columna "duration" para obtener el df resultante df_ej8

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
df_duracion = df_catalogo[df_catalogo["Runtime"].between(90, 150, inclusive = "both")]
df_ej9 = df_duracion[df_duracion["type"]== "Movie"].filter(items = ["title", "Runtime"])
df_ej9
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
patron_regex = "love"
filtro_love = df_catalogo["title"].str.contains(patron_regex, case = False)
filtro_TVshow = df_catalogo["type"] == "TV Show"
df_TVShow_Love = df_catalogo[filtro_love & filtro_TVshow]
df_TVShow_Love.filter(["title", "type"])
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
patron_regex2 = ".*Johnson(,|$)"
## Como es el apellido, ponemos .* para indicar que antes puede ir cualquier caracter y ponemos (,|$) para indicar que después puede ir una coma si hay más de un director o $ para indicar que es el final
filtro_director = df_catalogo["director"].str.contains(patron_regex2, na = False, case = False)
## na=False para que excluya los resultados nulos y case = False para que busque sin distinguir mayúsculas y minúsculas
df_catalogo[filtro_director]
```
